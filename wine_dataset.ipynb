```python
#fixing missing values for sulphates
meanSulphate = data["sulphates"].mean()
meanSulphate
data["sulphates"] = data["sulphates"].fillna(meanSulphate)
```

```python
#fixing missing values for ph
meanPh = data["pH"].mean()
meanPh
data["pH"] = data["pH"].fillna(meanPh)
```

```python
#fixing missing values for chlorides
meanChlorides = data["chlorides"].mean()
meanChlorides
data["chlorides"] = data["chlorides"].fillna(meanChlorides)
```

```python
#fixing missing values for residual sugar
residualSugar = data["residual sugar"].mean()
residualSugar
data["residual sugar"] = data["residual sugar"].fillna(residualSugar)
```

```python
#fixing missing values for citric acid
citiricAcid = data["citric acid"].mean()
citiricAcid
data["citric acid"] = data["citric acid"].fillna(citiricAcid)
```

```python
#fixing missing values for volatile acidity
volatileAcidity = data["volatile acidity"].mean()
volatileAcidity
data["volatile acidity"] = data["volatile acidity"].fillna(volatileAcidity)
```

```python
#fixed missing values for volatile acidity
fixedAcidity = data["fixed acidity"].mean()
fixedAcidity
data["fixed acidity"] = data["fixed acidity"].fillna(fixedAcidity)
```

```python
#convert the non numeric data to numeric
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ["type"]:
  le.fit(data[col])
  data[col] = le.transform(data[col])
```

```python
#using Sklearn library split the data into train and set set with the ration of 25% for test and 75% for training.
from sklearn.model_selection import train_test_split
#X :get all input data needed to predict output or outcome
X = data.iloc[:, 1:-1].values

#y get target column
y = data.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
#scale data here
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
# scaler = MinMaxScaler()
scaler = StandardScaler()
scaler.fit(data)
data = scaler.transform(data)

print("Min ", data.min())
print("Max ", data.max())
```

```python
#Classify the data once using LogisticRegression and once with KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
#Knearest neigbor classification
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=10)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
